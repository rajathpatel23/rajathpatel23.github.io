```python
# - dice A and dice B 
# - one play of the game with same two dice

import random

A = [9, 9, 9, 9, 0, 0]
B = [3, 3, 3, 3, 11, 11]

def roll_dice(A, B):
    diceA = random.choice(A)
    diceB = random.choice(B)
    if diceA > diceB:
        return "A_win"
    else:
        return "B_win"
```

```python
count_A = 0
count_B = 0
for _ in range(10000000):
    val = roll_dice(A, B)
    if val == 'A_win':
        count_A +=1
    else:
        count_B +=1
prob_A = count_A /10000000
```
